```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
ndvi = np.genfromtxt('ndvi20012012.csv', delimiter=',', skip_header=1)
ndvi = ndvi[:,1:]
```

```python
lst = np.genfromtxt('lst20012012.csv', delimiter=',',skip_header=1)
lst = lst[:,1:]
```

```python
nl = np.genfromtxt('night_lights20012012.csv', delimiter=',',skip_header=1)
nl = nl[:, 1:]
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.despine(left=True, bottom=True)
ax = sns.heatmap(ndvi[:,:], cbar_kws={'label': 'Average NDVI Index Change (2012 - 2002)'})
ax.set_xticks([]);
ax.set_yticks([]);
plt.savefig('NDVI.pdf');
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.despine(left=True, bottom=True)
ax = sns.heatmap(lst, cbar_kws={'label': 'Average LST Index Change (2012 - 2002)'})
ax.set_xticks([]);
ax.set_yticks([]);
plt.savefig('LST.pdf');
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.despine(left=True, bottom=True)
ax = sns.heatmap(nl, cbar_kws={'label': 'Average NightLight Index Change (2012 - 2002)'})
ax.set_xticks([]);
ax.set_yticks([]);
```
