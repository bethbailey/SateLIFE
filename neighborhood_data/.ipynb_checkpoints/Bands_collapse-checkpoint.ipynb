```python
import pickle
import numpy as np
import pandas as pd
```

```python
stds_data = {}
years = [2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012]
for year in years:
    with open('stds_by_neighborhood{}.cp'.format(year), 'rb') as fp:
        stds_data[year] = pickle.load(fp)
```

```python
mean_data = {}
for year in years:
    with open('means_by_neighborhood{}.cp'.format(year), 'rb') as fp:
        mean_data[year] = pickle.load(fp)
```

```python
min_data = {}
for year in years:
    with open('mins_by_neighborhood{}.cp'.format(year), 'rb') as fp:
        min_data[year] = pickle.load(fp)
```
